```python
import geopandas as gpd
import pandas as pd
from tqdm import tqdm
```

```python
access = gpd.read_file('/Users/zawoon96/Desktop/ST_ICU_Access/data/plots/ICU_access_county.shp')
access.set_index('index', inplace=True)
access = access.to_crs(epsg=3081)
access
```

```python
county = access.copy(deep=True)
county = county[['geometry']]
county
```

```python
od_matrix = pd.DataFrame(index=county.index, columns=county.index)
od_matrix
```

```python
for i in tqdm(county.index):
    for j in county.index:
        if i != j:
            dist = county.at[i, 'geometry'].centroid.distance(county.at[j, 'geometry'].centroid)
            od_matrix.at[i, j ] = dist
```

```python
od_matrix.min().quantile(0.90)
```
